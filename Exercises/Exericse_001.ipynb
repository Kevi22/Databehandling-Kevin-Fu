```python
import pandas as pd
autos = pd.read_json("../Data/autos.json")
autos
```

```python
jaguar_cars = autos[autos["make"] == "jaguar"]
autos.query("make == 'jaguar'")
jaguar_count = len(jaguar_cars)
most_expensive_jaguar_price = jaguar_cars['price'].max()

print(f"Number of Jaguar cars: {jaguar_count}")
print(f"Price of the most expensive Jaguar car: {most_expensive_jaguar_price:}")
```

```python
toyota_cars = autos[autos["make"] == "toyota"]
autos.query("make == 'toyota'")
toyota_count = len(toyota_cars)
most_expensive_toyota_price = toyota_cars["price"].max()
print(f"Number of Jaguar cars: {toyota_count}")
print(f"Price of the most expensive Jaguar car: {most_expensive_toyota_price:}")

```
